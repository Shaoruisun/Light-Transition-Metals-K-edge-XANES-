```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sp = {"Mo": [20016], "Ru": [22135], "Rh": [23238], "Y": [17050], "Ni": [8333], "Ti": [4965], "Nb": [18998], "Cr": [5990, 5999], "Co": [7710], "Cu": [8984], "Mn": [6539], "Zr": [18010], "Cd": [26739], "Zn": [9661], "Pd": [24371], "Fe": [7112, 7121], "Tc": [21062], "Ag": [25538], "Sc": [4490], "V": [5465]}


data = pd.read_pickle("./predict_result.pk")
data.head()
```

```python
data_id = 0


start = int(sp[data.index.to_list()[data_id].split("-")[-3]][0])
x = [i for i in np.arange(start=start, stop=start+55.5, step=0.5)]
y = data.iloc[data_id,:]

plt.figure(figsize=(15,8),dpi=100)
plt.plot(x, y, color='b', label='Pre Data', linewidth=3)
plt.ylim((0,max(y)+0.3))

tickssize=25
plt.xticks(fontsize=tickssize)
plt.yticks(fontsize=tickssize)
labelsize=25
plt.xlabel('Energy/eV',fontsize=labelsize,color='black')
plt.ylabel('Intensity/a.u.',fontsize=labelsize,color='black')
plt.title(label=data.index.to_list()[data_id],fontsize=30,loc='center',color='black')
plt.rcParams.update({'font.size':25})
plt.legend()
plt.show()
```
